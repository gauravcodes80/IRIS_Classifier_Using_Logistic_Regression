```python
# Import necessary libraries
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Load the iris dataset
iris = datasets.load_iris()

# Print the keys of the dataset to understand its structure
print(iris.keys())

# Select the petal width feature (3rd index of data)
X = iris["data"][:, 3:]

# Create binary target variable: True if class is Virginica (label 2), False otherwise
Y = (iris["target"] == 2)

# Train a logistic regression model to predict whether a flower is Virginica or not
clf = LogisticRegression()
clf.fit(X, Y)

# Predict whether a flower with 1.6 cm petal width is Virginica (True/False)
predict = clf.predict([[1.6]])
print(predict)

# Generate a range of petal widths for plotting
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)

# Predict probability that the flower is Virginica for each petal width
Y_prob = clf.predict_proba(X_new)

# Plot the probability curve for Virginica (class 1)
plt.plot(X_new, Y_prob[:, 1], "g-", label="Virginica")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Probability of being Virginica")
plt.title("Logistic Regression - Virginica Classifier")
plt.legend()
plt.show()
```
